```python
#Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Set options for libraries
pd.set_option('display.max_rows',100)
pd.set_option('display.max_columns',60)
sns.set(style="darkgrid")
```

```python
#Read csv
df = pd.read_csv("contribution_network_properties.csv")
```

```python
#Describe data / initial exploratory analysis
print(df.describe())
```

```python
df.corr()
```

```python
df2 = df
```

```python
df2.drop('id', axis=1, inplace=True)
df2.drop('pledge_id', axis=1, inplace=True)
df2.drop('contribution_id', axis = 1, inplace=True)
```

```python
f, ax = plt.subplots(figsize=(9, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.corrplot(df2, annot=False, sig_stars=False,diag_names=False, cmap=cmap, ax=ax)
f.tight_layout()
```
